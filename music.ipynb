```python
sns.countplot(y = "Genre", data = music, palette="PuBu")
#alternative/indie is present the most number of times
```

```python
g = ("alternative ; indie ; electro", "alternative ; indie", "alternative ; indie ; pop", "alternative ; indie ; blues", "alternative ; indie ; pop ; rock")
```

```python
count = 0
for i in music["Genre"]:
    if i in g:
        count = count + 1
print(count)
#there is 19 songs out of 27 that are alternative/indie
```
